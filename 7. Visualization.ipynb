# Data Visualization with matplotlib

```python
# Import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## Global API in matplotlib (inspired by MATLAB style)

```python
# create numpy list
x = np.arange(-10, 11)
x
```

```python
# Create simple plotting with plt
plt.figure(figsize=(20,10))
plt.title('Simple Plot')
plt.plot(x, x**2)
plt.plot(x, -1 *(x**2))
```

```python
plt.figure(figsize=(20,10))
plt.title('Simple Plot')

plt.subplot(1, 2, 1)
plt.plot(x, x**2)
plt.plot([0,0,0],[-10, 0, 100])
plt.legend(['X^2', 'Vertical Line'])
plt.xlabel('X')
plt.ylabel('X Squared')

plt.subplot(1, 2, 2)
plt.plot(x, (x**2)*-1)
plt.plot([-10,0,10],[-50, -50, -50])
plt.legend(['-X^2', 'Vertical Line'])
plt.xlabel('X')
plt.ylabel('X Squared')
```

## OOP approcah (Reference each figure or axes by its own object)

```python
fig, axes = plt.subplots(figsize=(20, 10))
```

```python
fig
```

```python
axes.plot(
    x, (x ** 2), color='red', linewidth=3, marker='o', markersize=8, label='X^2'
)
axes.plot(
    x, (x ** 2)*-1, 'b--', linewidth=3, marker='o', markersize=8, label='-X^2'
)
axes.set_xlabel('X')
axes.set_ylabel('X Squared')

axes.set_title('Basic X Square Plot')
axes.legend()

fig
```

```python
# Create new plot
fig, axes = plt.subplots(figsize=(20, 10))

# Different line styles
axes.plot(x, 2*x + 0, linestyle='solid')
axes.plot(x, 2*x + 1, linestyle='dashed')
axes.plot(x, 2*x + 2, linestyle='dashdot')
axes.plot(x, 2*x + 3, linestyle='dotted')

axes.set_title("Line Plotting")
```

## Figures aand subfigures

```python
plot_objects = plt.subplots(nrows=2, ncols=2, figsize=(20, 10))

fig, ((ax1, ax2), (ax3, ax4)) = plot_objects

plot_objects
```

```python
# 4 random plot
ax4.plot(np.random.randn(50), c='yellow')
ax1.plot(np.random.randn(50), c='red', linestyle='--')
ax2.plot(np.random.randn(50), c='green', linestyle=':')
ax3.plot(np.random.randn(50), c='blue', marker='o', linewidth=3.0)

ax1.set_title("Random plot 1")
ax2.set_title("Random plot 2")
ax3.set_title("Random plot 3")
ax4.set_title("Random plot 4")

fig
```

## Scatter Plot

```python
# Preparing Variable
N = 100
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (200 * np.random.rand(N)**2)
```

```python
plt.figure(figsize=(20,10))

# Creating random scatter plot
plt.scatter(x, y)
plt.title("Random Scatter Plot")
plt.show()
```

```python
plt.figure(figsize=(20,10))

# Creating another scatter plot
plt.scatter(x, y, s=area, c=colors, cmap='Spectral', alpha=0.5)
plt.title("Another Random Scatter Plot")
plt.colorbar()
plt.show()
```

## Histogram

```python
# Preparing values
values = np.random.randn(1000)
```

```python
plt.figure(figsize=(20,10))

plt.hist(values)
```

```python
plt.figure(figsize=(20,10))

plt.hist(values, bins=100, histtype='bar', color='blue', edgecolor='green', alpha=0.7)
plt.title('Random Histogram Figure')
plt.xlim(xmin=-5, xmax=5)
```

## Barplot

```python
# Preparing variable
y = np.random.rand(1, 5)[0]
y2 = np.random.rand(1, 5)[0]
```

```python
plt.figure(figsize=(20, 10))

barWidth = 0.5
plt.bar(np.arange(len(y)), y, width=barWidth, color='#00b894')

plt.show()
```

```python
plt.figure(figsize=(20, 10))
barWidth = 0.5
plt.bar(np.arange(len(y)), y, width=barWidth, color='#00b894', label='Label Y')
plt.bar(np.arange(len(y2)), y2, width=barWidth, color='#e17055', bottom=y, label='Label Y2')

plt.title('Random Bar plot')
plt.legend()
plt.show()
```
