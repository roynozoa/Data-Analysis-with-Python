## Customize seaborn with matplotlib object

```python
fig, ax = plt.subplots()
sns.distplot(movie_df['title_year'], rug=True, bins=20, ax=ax)
ax.set(xlabel='Year', ylabel='Distribution', xlim=(1985,2025), title='Movie Distribution by Year')
```

```python
fig, ax = plt.subplots()
sns.distplot(movie_df['gross'], bins=15, ax=ax)
ax.set(xlabel='gross', ylabel='Distribution', title='Movie Distribution by gross')

ax.axvline(x=movie_df['gross'].median(), color='m', label='Median', linestyle='--', linewidth=2)
ax.axvline(x=movie_df['gross'].mean(), color='b', label='Mean', linestyle='-', linewidth=2)


ax.legend()
plt.show()
```

```python
flights = sns.load_dataset('flights', parse_dates=[0])
flights['year'] = flights['year'].dt.year
flights = flights.set_index('year')
flights.head()
```

```python
sns.regplot(data=flights, x=flights.index, y='passengers', order=2, x_jitter=.1, x_estimator=np.median, x_bins=20)
```

```python
sns.regplot(data=flights, x=flights.index, y='passengers')
```

```python
# matrix plots
# sns.heatmap() and pd.crosstab()
flight_tab = pd.crosstab(flights['month'], flights.index, values=flights['passengers'],aggfunc='mean').round(0)
flight_tab
```

```python
sns.heatmap(flight_tab)
```

```python
sns.heatmap(flight_tab, annot=True, fmt='d', cbar=False, linewidths=.5)
```

```python
sns.heatmap(flight_tab.corr())
```

# Plots on Data Aware Grids

 ### FacetGrid, factorplot, and lmplot
 also called faceting
 factorplot and lmplot are the shortcut for FacetGrid

```python
plt.figure(figsize=(12,8))
sns.factorplot(y='passengers',col='month', data=flights, kind='box', col_wrap=3)
```

```python
fg = sns.FacetGrid(data=movie_df,col='genre', col_wrap=3)
fg.map(plt.scatter, 'budget', 'gross')
```

```python
sns.lmplot(data=movie_df, x='budget', y='gross', row='genre', col='country')
```

### PairGrid and pairplot
pairplot is a shortcut for PairGrid

```python
g = sns.PairGrid(movie_df, vars=["budget", "gross"])
g2 = g.map_diag(plt.hist)
g3 = g2.map_offdiag(plt.scatter)
```

```python
sns.pairplot(movie_df, vars=["budget", "gross"], kind='scatter', hue='genre', diag_kws={'alpha':.5})
```

### JointGrid and jointplot
jointplot is a shortcut for JointGrid

```python
g = sns.JointGrid(data=movie_df, x='budget', y='gross')
g.plot(sns.regplot, sns.kdeplot)
```

```python
g = sns.JointGrid(data=movie_df, x='budget', y='gross')
g = g.plot_joint(sns.kdeplot)
g = g.plot_marginals(sns.kdeplot, shade=True)
# g = g.annotate(stats.pearsonr)
```

```python
sns.jointplot(data=movie_df, x='budget', y='gross', kind='reg')
```

```python
sns.jointplot(data=movie_df, x='budget', y='gross', kind='hex')
```
