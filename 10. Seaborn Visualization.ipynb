# Seaborn for data visualization

## Seaborn
- A library in python for data visualization
- Can create common type of plots easily  
- Very useful and easy to use
- Works well with pandas data structures
- Seaborn was buit with matplotlib
- In order to use seaborn library, matplotlib is also needed

```python
# import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# simple list of height and weight
weight = [65, 65, 70, 80, 56, 45, 76, 66, 61, 62]
height = [141, 155, 161, 180, 155, 151, 183, 176, 155, 170]
```

```python
help(sns.scatterplot)
```

```python
# create a basic scatter plot with seaborn
sns.scatterplot(x=weight, y=height)
```

```python
# regplot
sns.regplot(x=weight, y=height)
```

```python
# countplot
Sex = ['M', 'M', 'F', 'M', 'F', 'F', 'F', 'M', 'M', 'F', 'F', 'F', 'F']
sns.countplot(x=Sex)
```

## Using pandas with Seaborn 

```python
game = pd.read_csv('data/csv/GeneralEsportData.csv',  encoding = "ISO-8859-1", index_col=0, parse_dates=[1])
```

```python
game['ReleaseDate'] = game['ReleaseDate'].dt.year
```

```python
game.info()
```

```python
game.head()
```

```python
plt.figure(figsize=(20,10))
game_plot = sns.relplot(x='Genre', y='TotalPlayers', kind='line', data=game, col='ReleaseDate', col_wrap=3, ci=None)
game_plot.set_xticklabels(rotation=90)
game_plot.fig.suptitle('Number of players for each video game genre in 1981 - 2020 ', y=1.01)
```

```python
game_plot2 = sns.catplot(x='Genre', y='TotalPlayers', kind='bar', data=game, col='ReleaseDate', col_wrap=3, ci=None)
game_plot2.set_xticklabels(rotation=90)
game_plot2.fig.suptitle('Number of players for each video game genre in 1981 - 2020 ', y=1.01)
```

```python
sns.relplot(x='ReleaseDate', y='TotalTournaments', kind='line', data=game, hue='Genre', ci=None)
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x='ReleaseDate', y='TotalEarnings', data=game)
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x='Genre', data=game)
plt.xticks(rotation=90)
```

## Line plot using relplot()

```python
flights = sns.load_dataset('flights', parse_dates=[0])
flights.head()
```

```python
flights.info()
```

```python
flights['year'] = flights['year'].dt.year
```

```python
flights.head()
```

```python
flights = flights.set_index('year')
```

```python
sns.relplot(x='month', y='passengers', kind='line', data=flights, col=flights.index, col_wrap=2, ci='sd')
```

```python
sns.relplot(x='month', y='passengers', kind='line', data=flights, style=flights.index, hue=flights.index, markers=True, dashes=False)
```

```python
sns.relplot(x='month', y='passengers', kind='line', data=flights, ci='sd')
```

# Categorical plot with catplot()

Count Plots, Bar plots, Box plots, and Point plots with catplot()

```python
sns.catplot(kind='bar', x=flights.index, y='passengers', data=flights)
```

```python
# box plot
sns.catplot(kind='box', x=flights.index, y='passengers', data=flights, whis=[5, 95])
```

```python
# point plots
sns.catplot(kind='point', x=flights.index, y='passengers', data=flights, estimator=np.median, capsize=0.2)
```

# Styling Seaborn Figure

- sns.set_style() Preset options: 'white'(default), 'dark', 'whitegrid', 'darkgrid', 'ticks'
- sns.color_palette() Preset options : 'RdBu', 'PRGn', 'RdBu_r', 'PRGn_r', 'Greys', 'Blues', 'PuRd', 'GnBu'
- sns.set_context() Preset options : 'paper', 'notebook', 'talk', 'poster'

```python
plt.figure(figsize=(20,10))

sns.set_style('darkgrid')

sns.set_context('notebook')
sns.catplot(kind='bar', x=flights.index, y='passengers', data=flights, ci=None)

sns.set_palette(['#39A7D0', "#36ADA4"])
```

## Adding titles and labels

```python
# box plot
box_plot = sns.catplot(kind='box', x=flights.index, y='passengers', data=flights, whis=[5, 95])

line_plot = sns.relplot(x='month', y='passengers', kind='line', data=flights, style=flights.index, hue=flights.index, markers=True, dashes=False)

box_plot.fig.suptitle('Box Plot', y=1.03)
line_plot.fig.suptitle('Line Plot', y=1.03)
```

```python
# axessubplot
sns.set_style('darkgrid')
sns.set_palette(['#39A7D0', "#36ADA4"])
plot = sns.relplot(x='month', y='passengers', kind='line', data=flights, col=flights.index, col_wrap=2)
plot.fig.suptitle('Number of Flight Passengers im 1949 - 1960', y=1.01)
plot.set_titles('Year {col_name}')
plt.show()
```
